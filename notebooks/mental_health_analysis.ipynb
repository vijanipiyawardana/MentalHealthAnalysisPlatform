### Building Deep Learning Model with fastai

```python
from fastai.tabular.all import *
```

```python
# Define categorical and continuous columns

categorical_cols = ['IsSubstanceReported', 'IsAlcohol', 'IsMarijuana', 'IsOpium', 'IsHeroin', 
                    'IsHygieneAppropriate', 'IsPostureAppropriate', 'IsBehaviourCooperative', 'IsMotorPsycho', 'ISGroomingAppropriate',
                    'ISSpeechPressure', 'ISSpeechNormal', 'ISSpeechRapid', 'ISSpeechSlow', 'ISSpeechWhispered', 'ISSpeechLoud', 'ISSpeechSpontaneous', 'ISSpeechMute', 'ISSpeechShuttering', 'ISSpeechContentRelevant',
                    'ISObjectiveEuthymia', 'ISObjectiveAngry', 'ISObjectiveEuphoric', 'ISObjectiveApathetic', 'ISObjectiveDysphoric', 'ISObjectiveLabile',
                    'ISThoughtCircumstantiality', 'ISThoughtFlight', 'ISThoughtLoose', 'ISThoughtTangentiality',
                    'ISContentHomicidal', 'ISContentMagical', 'ISContentObsessious', 'ISContentOvervaluedIdeas', 'ISContentParanoia', 'ISContentPhobic', 'ISContentPovertySpeech', 'ISContentPreoccupations', 'ISContentSuicidalIdeation', 'ISContentSuspiciousness',
                    'ISPreceptionVisual', 'ISPreceptionAuditory', 'ISPreceptionTactile', 'ISPreceptionOlfactory', 'ISPreceptionReported', 'ISDepersonalizationReported', 'ISTimeIntact', 'ISPlaceIntact', 'ISPersonIntact', 'ISAttentionForwardIntact', 'ISAttentionBackwardIntact', 'ISMemoryImmediateIntact', 'ISMemoryShortTermIntact', 'ISMemoryLongTermIntact', 'ISAbstractThinkingIntact',
                    ]

continuous_cols = ['LowerBP', 'RespiratoryRate', 'Temp', 'Weight', 'Height', 'UpperBP', 'OnsetAge', 'Episode', 'Duration', 
                   'UseDuration', 'DoseFrequency']
```

```python
df['Diagnosis'] = df['Diagnosis'].astype('category')
```

```python
# Split the dataset into training and validation sets
#splits = RandomSplitter(valid_pct=0.2)(range_of(df))
train_idx = df.sample(frac=0.7, random_state=42).index  # 70% for training
remaining_idx = df.drop(train_idx).index
valid_idx = remaining_idx[:int(0.67 * len(remaining_idx))]  # 20% for validation
test_idx = remaining_idx[int(0.67 * len(remaining_idx)):]  # 10% for testing

#valid_idx = df.drop(train_idx).index  # 20% for validation
splits = (list(train_idx), list(valid_idx))

# Specify dependent variable (target) and independent variables (features)
target_col = 'Diagnosis'
#features = df.columns[df.columns != target].tolist()
```

```python
# Create a TabularPandas object

to = TabularPandas(
    df,
    procs=[Categorify, FillMissing, Normalize],
    cat_names=categorical_cols,
    cont_names=continuous_cols,
    y_names=target_col,
    splits=splits
)

#cont_names = features  # Continuous variables
#procs = [Categorify, FillMissing, Normalize]  # Data transformation steps
#to = TabularPandas(df, procs, cont_names, target, splits=(train_idx, valid_idx))
```

### Evaluate model on the test set

```python
# Filter the test data directly from the original DataFrame
test_data = df.iloc[test_idx]
```

```python
# Create a TabularPandas object for the test data
test_to = TabularPandas(
    test_data,
    procs=[Categorify, FillMissing, Normalize],
    cat_names=categorical_cols,
    cont_names=continuous_cols,
    y_names=target_col
)
```

```python
# Create a DataLoader for the test data
test_dls = test_to.dataloaders(bs=64)
```

```python
# Get predictions on the test data
predictions, _ = learn.get_preds(dl=test_dls.train)
```

```python
# Decode predictions
predicted_classes = predictions.argmax(dim=1)
actual_classes = test_data[target_col].values
```

```python
# Initialize a counter for correct predictions
correct_count = 0

# Compare predictions with actual classes
for i in range(len(test_data)):
    actual_class = actual_classes[i]
    predicted_class = dls.vocab[int(predicted_classes[i])]  # Use dls.vocab instead of test_to.classes
    is_correct = actual_class == predicted_class
    correct_count += int(is_correct)  # Increment counter if prediction is correct
    print(f"Sample {i + 1}: Actual: {actual_class}, Predicted: {predicted_class}, Correct: {is_correct}")

# Calculate percentage of correct predictions
accuracy_percentage = (correct_count / len(test_data)) * 100
print(f"Percentage of Correctly Predicted Samples: {accuracy_percentage:.2f}%")
```
