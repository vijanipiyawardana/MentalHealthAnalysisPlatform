### Feature Importance by Permutation

```python

import torch
import pandas as pd
import numpy as np
from sklearn.metrics import accuracy_score

# Define the FastAI model wrapper
class FastAIModelWrapper:
    def __init__(self, learner):
        self.learner = learner

    def predict(self, X):
        # Ensure X is in DataFrame format with column names
        if isinstance(X, np.ndarray):
            X = pd.DataFrame(X, columns=self.learner.dls.x_names)

        # Create a test DataLoader and get predictions
        dl = self.learner.dls.test_dl(X)
        preds, _ = self.learner.get_preds(dl=dl)
        return preds.argmax(dim=1).numpy()  # Return predicted class indices

# Wrap the FastAI learner
wrapped_model = FastAIModelWrapper(learn)
```

```python

# Function to compute accuracy
def compute_accuracy(model, X, y):
    preds = model.predict(X)
    return accuracy_score(y, preds)
    
```

```python

# Manual permutation importance function
def permutation_importance_manually(model, X, y, n_repeats=10, random_state=42):
    baseline_accuracy = compute_accuracy(model, X, y)
    importances = np.zeros(X.shape[1])
    np.random.seed(random_state)

    for i in range(X.shape[1]):
        shuffled_X = X.copy()
        for _ in range(n_repeats):
            np.random.shuffle(shuffled_X[:, i])
            accuracy = compute_accuracy(model, shuffled_X, y)
            importances[i] += baseline_accuracy - accuracy
    importances /= n_repeats
    return importances
    
```

```python

# Prepare validation data
X_valid = to.valid.xs  # Feature matrix (DataFrame)
y_valid = to.valid.y.to_numpy()  # Target labels (numpy array)

# Run permutation importance
importances = permutation_importance_manually(
    model=wrapped_model,
    X=X_valid.to_numpy(),
    y=y_valid,
    n_repeats=10,
    random_state=42
)
```

```python
# Print feature importances


# Combine features and their importances
feature_importances = list(zip(X_valid.columns, importances))

# Sort features by importance (descending order)
sorted_importances = sorted(feature_importances, key=lambda x: x[1], reverse=True)

# Print sorted feature importances
print("Feature Importances (Sorted by Mean Decrease in Accuracy):")
for feature, importance in sorted_importances:
    print(f"{feature}: {importance}")
    
```
